This notebook allows you to scan a list of email addresses and check if they have been exposed in public data breaches using the HaveIBeenPwned API.

```python
import pandas as pd
from src.hibp_lookup import check_bulk_emails

API_KEY = "your_api_key_here"
```

```python
# Load email list
df = pd.read_csv("../data/emails.csv")
email_list = df["email"].dropna().tolist()
email_list
```

```python
# Run breach checks
results = check_bulk_emails(email_list, API_KEY)

# Format results
summary = []
for email, breaches in results.items():
    summary.append({
        "email": email,
        "breaches_found": len(breaches) if breaches else 0,
        "breaches": [b['Name'] for b in breaches] if breaches else []
    })

result_df = pd.DataFrame(summary)
result_df
```
